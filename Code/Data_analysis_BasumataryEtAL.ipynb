```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib qt5
```

```python
df = pd.read_csv('Data_BasumataryEtAl.csv')
df.head(54)
```

```python
df_reaches_40to50 = df.iloc[39:50]
df_reaches_40to50
```

# Finding Percent Erosion in the Critical ROI ( 1987 - 2016 )

```python
net_erosion = df['Mean Erosion (ha)'].sum()
net_erosion
```

```python
net_erosion_40to50 = df_reaches_40to50['Mean Erosion (ha)'].sum()
net_erosion_40to50
```

```python
print("Percentage Erosion in Critical ROI (in %) = ", net_erosion_40to50*100/net_erosion)
```

# Finding Percent Change in the Critical ROI ( 1987 - 2016 )

```python
net_change = df['Net change(ha)'].sum()
net_change
```

```python
net_change_40to50 = df_reaches_40to50['Net change(ha)'].sum()
net_change_40to50
```

```python
print("Percentage Change in Critical ROI (in %) = ", net_change_40to50*100/net_change)
```
